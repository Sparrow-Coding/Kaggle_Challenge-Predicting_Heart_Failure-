# EDA

# Column = age
### Age minimum : 40
### Age maximum : 95
### Age average : 60



```python
sns.distplot(df['age'], bins = 50)
plt.xlabel('Age')
plt.ylabel('Count')
plt.axvline(x = df['age'].mean(), color = 'red', ls = '--', label='mean')
plt.axvline(x = df['age'].median(), color = 'black', ls = '--', label ='median')
plt.title('Distribution of age')
plt.legend()
plt.savefig('Distribution of age.png')
```
